## Association Rule

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

using https://www.kaggle.com/datasets/carrie1/ecommerce-data

```python
df = pd.read_csv("../data/e_commerce_kaggle/data.csv", encoding="unicode_escape")
```

```python
df = (
    df.dropna()
    .query("Quantity > 0")
    .assign(Amount=lambda _df: df.Quantity * df.UnitPrice)
)
```

```python
df
```

```python
df.InvoiceDate = pd.to_datetime(df.InvoiceDate, format="%m/%d/%Y %H:%M")
```

```python
df["year_month"] = df.InvoiceDate.apply(lambda x: 100 * x.year + x.month)
```

```python
df["month"] = df.InvoiceDate.dt.month
df["day"] = df.InvoiceDate.dt.day_of_week + 1
df["hour"] = df.InvoiceDate.dt.hour
```

```python
market_basket = (
    df.groupby(["InvoiceNo", "Description"])
    .Quantity.sum()
    .unstack()
    .reset_index()
    .fillna(0)
    .set_index("InvoiceNo")
)

market_basket
```

```python
def is_purchased(x):
    return 0 if x < 1 else 1
```

```python
market_basket = market_basket.applymap(is_purchased)
```

```python
market_basket
```

```python
def bought_together(product_invoice, product_name):
    product_df = product_invoice.loc[product_invoice[product_name] == 1]

    product_freq = apriori(product_df, min_support=0.15, use_colnames=True)

    product_rules = (
        association_rules(product_freq, metric="lift", min_threshold=1)
        .sort_values(by=["lift", "support"], ascending=False)
        .reset_index(drop=True)
    )

    return product_rules.consequents.unique()[:6]
```

```python
bought_together(market_basket, "WOODEN HEART CHRISTMAS SCANDINAVIAN")
```

```python
bought_together(market_basket, "WHITE METAL LANTERN")
```

```python
bought_together(market_basket, "JAM MAKING SET WITH JARS")
```
